```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
# import pyroj
```

```python
# set data paths
readpath = '../download/historical_archive'
writepath = '../download/historical_survey'
```

```python
# load plot locations 
plotlocations = pd.read_csv(f"{readpath}/RP.Plot_locations.csv")
```

```python
plotlocations.head()
```

```python
# create geopandas df with geometries from lat lon
plots = gpd.GeoDataFrame(
    plotlocations, 
    geometry=gpd.points_from_xy(plotlocations.LONGITUDE, plotlocations.LATITUDE), crs="EPSG:4326"
)
```

```python
type(plots)
```

```python
plots.head()
```

```python
# reproject to UTM grid
plots.to_crs(crs='EPSG:32617', inplace=True)
```

```python
# get first plot in first unit
plot=plots[(plots['UNIT'] == 1) & (plots['PLOT'] == 1)]
```

```python
plot
```

```python
print(plot.iloc[0].geometry.x,plot.iloc[0].geometry.y)
```

```python
# load tree locations
df = pd.read_csv(f"{readpath}/RP.Plot_data.csv")
print(df.head())
```

```python
# get all trees in unit 1 plot 1
subset=df[(df['UNIT'] == 1) & (df['PLOT'] == 1)].copy()
```

```python
# Print the subset DataFrame to verify its content
print(subset.head())
```

```python
# create geopandas df with geometries from x y coords
offset = 50
subset['XCOORD'] = subset['XCOORD'] + plot.iloc[0].geometry.x - offset
subset['YCOORD'] = subset['YCOORD'] + plot.iloc[0].geometry.y - offset
plot_trees = gpd.GeoDataFrame(
    subset, 
    geometry=gpd.points_from_xy(subset.XCOORD, subset.YCOORD), crs="EPSG:32617"
)
```

```python
# save to file
plot_trees.to_file(f"{writepath}/plot_1_1.shp", driver="ESRI Shapefile")
```
