```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
# import pyroj
```

```python
# set data path
path = '../download/historical_archive'
```

```python
# load plot locations 
plotlocations = pd.read_csv(f"{path}/RP.Plot_locations.csv")
```

```python
plotlocations.head()
```

```python
# create geopandas df with geometries from lat lon
plots = gpd.GeoDataFrame(
    plotlocations, 
    geometry=gpd.points_from_xy(plotlocations.LONGITUDE, plotlocations.LATITUDE), crs="EPSG:4326"
)
```

```python
type(plots)
```

```python
plots.head()
```

```python
# reproject to UTM grid
plots.to_crs(crs='EPSG:32617', inplace=True)
```

```python
# get first plot in first unit
plot=plots[(plots['UNIT'] == 1) & (plots['PLOT'] == 1)]
```
