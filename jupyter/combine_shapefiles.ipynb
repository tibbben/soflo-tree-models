# Combine Shapefiles into GeoPackage

This notebook combines all shapefiles in the `historical_survey` folder into a single GeoPackage file for use in visualization and model training.

## Requirements
The following Python packages must be installed in your environment:
- `geopandas`
- `pandas`

To install them run the following in your terminal:
```
pip install geopandas pandas
```

````python
import geopandas as gpd
import pandas as pd
import glob
````

````python
# path to the folder containing all shapefiles
folder = "./download/historical_survey"

# find all shapefiles recursively in the folder
shapefiles = glob.glob(folder + "/**/*.shp", recursive=True)

print("Found", len(shapefiles), "shapefiles.")
````

````python
# read each shapefile into a list
all_dataframes = []
for shp in shapefiles:
    gdf = gpd.read_file(shp)
    all_dataframes.append(gdf)

# stack all dataframes into one
merged = pd.concat(all_dataframes, ignore_index=True)

# convert to GeoDataFrame
combined = gpd.GeoDataFrame(merged)
````

````python
# save to GeoPackage
output = "./download/combined_trees.gpkg"
combined.to_file(output, driver="GPKG")

print("Done! Combined", len(shapefiles), "shapefiles into combined_trees.gpkg")
````
