```python
folder = r"C:\Users\ahsan\Documents\Ahsan\UMiami\GDSC Tree Training\big_cypress\download"

shapefiles = glob.glob(folder + "\\**\\*.shp", recursive=True)
print(f"Found {len(shapefiles)} shapefiles:")
for shp in shapefiles:
    print(shp)
```

```python
# read each shapefile into a list
all_dataframes = []
for shp in shapefiles:
    gdf = gpd.read_file(shp)
    all_dataframes.append(gdf)

# stack them all into one big table
merged = pd.concat(all_dataframes, ignore_index=True)

# convert to a GeoDataFrame
combined = gpd.GeoDataFrame(merged)
```

```python
output = folder + "\\combined_trees.gpkg"
combined.to_file(output, driver="GPKG")
```
